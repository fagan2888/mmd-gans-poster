```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set(context='poster', style='whitegrid', font_scale=1.5)
```

```python
def r(l):
    return [l[i] for i in [1, 2, 0, 3]]

names    = r(["MMD RQ*", "Cram√©r", "WGAN-GP", "test set"])
FIDs     = r([20.55, 31.30, 29.24, 2.25])
FID_stds = r([  .25,   .17,   .22,  .04])
KIDs     = r([.013, .025, .022, 0])
KID_stds = r([.001, .001, .001, 0])

lsun_FIDs     = r([31.95, 54.18, 41.39, 2.49])
lsun_FID_stds = r([  .28,   .39,   .25,  .02])
lsun_KIDs     = r([.028, .050, .039, 0])
lsun_KID_stds = r([.002, .002, .002, 0])
```

```python
c1, c2 = sns.color_palette(n_colors=2)
inds = np.arange(len(names))
width = .35
```

```python
fig = plt.figure(figsize=(5, 3))
ax = plt.gca()
ax2 = ax.twinx()

ax .bar(inds,         FIDs, yerr=FID_stds, color=c1, width=width)
ax2.bar(inds + width, KIDs, yerr=KID_stds, color=c2, width=width)
ax.set_ylabel('FID')
ax2.set_ylabel('KID')
ax.set_xticks(inds + width/2)
ax.set_xticklabels(names, rotation=20, ha='center')

ax.set_yticks([0, 10, 20, 30])
ax.set_ylim(0, 35)
ax2.set_ylim(0, .035)
ax2.set_yticks([0, .01, .02, .03])
ax2.grid(False)

ax.set_title('CelebA')
fig.savefig('figs/celeba-scores.pdf', bbox_inches='tight', pad_inches=0)
```

```python
fig = plt.figure(figsize=(5, 3))
ax = plt.gca()
ax2 = ax.twinx()

ax .bar(inds,         lsun_FIDs, yerr=lsun_FID_stds, color=c1, width=width)
ax2.bar(inds + width, lsun_KIDs, yerr=lsun_KID_stds, color=c2, width=width)
ax.set_ylabel('FID')
ax2.set_ylabel('KID')
ax.set_xticks(inds + width/2)
ax.set_xticklabels(names, rotation=20, ha='center')

ax.set_ylim(0, 60)
ax2.set_ylim(0, .06)
ax2.grid(False)

ax.set_title('LSUN bedrooms')

fig.savefig('figs/lsun-scores.pdf', bbox_inches='tight', pad_inches=0)
```
