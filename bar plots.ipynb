```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set(context='poster', style='white')
```

```python
names = ["MMD RQ*", "Cram√©r", "WGAN-GP", "test set"]
FIDs = [20.55, 31.30, 29.24, 2.25]
FID_stds = [.25, .17, .22, .04]
KIDs = [.013, .025, .022, 0]
KID_stds = [.001, .001, .001, 0]
```

```python
fig, (axFID, axKID) = plt.subplots(1, 2, sharex=True, )

cFID, cKID = sns.color_palette(n_colors=2)
inds = np.arange(len(names))

axFID.bar(inds, FIDs, yerr=FID_stds, color=cFID)
axKID.bar(inds, KIDs, yerr=KID_stds, color=cKID)

axKID.set_xticks(inds)
axKID.set_xticklabels(names)
plt.tight_layout()
```
